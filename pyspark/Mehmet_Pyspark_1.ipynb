```python
from pyspark.ml.feature import StringIndexer
```

```python
indexer=StringIndexer(inputCols=['Rank','Country_Name','Male_Height_in_Cm','Female_Height_in_Cm','Male_Height_in_Ft','Female_Height_in_Ft'],
                      outputCols=['rank','countryname','maleheightincm','femaleheightincm','maleheightinft','femaleheightinft'])
df_r=indexer.fit(dataframe).transform(dataframe)
```

```python
from pyspark.sql.types import IntegerType
```

```python
from pyspark.sql.dataframe import DataFrame
data = dataframe.withColumn("Rank", dataframe["Rank"].cast(IntegerType()))
data = data.withColumn("Male_Height_in_Cm", data["Male_Height_in_Cm"].cast(IntegerType()))
data = data.withColumn("Female_Height_in_Cm", data["Female_Height_in_Cm"].cast(IntegerType()))
data = data.withColumn("Female_Height_in_Ft", data["Female_Height_in_Ft"].cast(IntegerType()))
```
